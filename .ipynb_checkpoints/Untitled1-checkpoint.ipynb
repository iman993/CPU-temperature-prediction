# Prédire la température de CPU Package

Prédire la température de votre système en fonction de facteurs tels que l'utilisation de la RAM, le Total du CPU utilisé

```python
#libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#loading the dataset
dataset = pd.read_csv("temperature.csv")
dataset.head()
```

```python
#librairie graphique
import seaborn as sns
#pairplot
sns.pairplot(dataset)
```

```python
#removing cpu temp and Memory  from dataset
dataset.head()
dataset.dtypes
ds  = dataset.drop(columns = ['CPU Package Temperature','Memory'])
print(ds)
```

# cas d'une seul dimension

# 1 CPU package Temperature en fonction de Total CPU

```python
# Save the data into variables
X = ds['CPU Total'].to_numpy()
# Each value in it's own array
X = X.reshape(-1, 1)
#output feature
Y = dataset.iloc[:,1].values
```

```python
#graphique avec les échelles carrées
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(X,Y,'o')
ax.set_xlabel("variable explicative :Total CPU")
ax.set_ylabel("variable expliquée ;temperature de CPU Package")
plt.title("visualisation")
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
# default test set size = 0.25 * data
X_train, X_test, y_train, y_test = train_test_split(X, Y,random_state = 0)
# Read data into dataframes
X_train_df, X_test_df = pd.DataFrame(X_train), pd.DataFrame(X_test)
print("train size: ", X_train.size)
print("validation size: ", X_test.size)
```

```python
from sklearn.model_selection import cross_val_score
E_test=[]
for i in range(10):
    polyreg1D = PolynomialFeatures(degree =(i+1)) # Transform X into X_Poly which contains X, X squared, X cubed, etc
    X_Poly = polyreg1D.fit_transform(X_train_df)
    regressor1 = LinearRegression() # Fit X_Poly in our Linear Regression model
    regressor1.fit(X_Poly,y_train)
    mset=-cross_val_score(regressor1,X_Poly,y_train,scoring="neg_mean_squared_error").mean()
    E_test.append(mset)
    print(mset)
```

```python
#proportion de variance expliquée
fig, ax = plt.subplots(figsize=(7,7))
ax.plot(range(1,11),E_test,".-")
ax.set_xlabel("degree de polynome")
ax.set_ylabel("erreur ")
plt.title("l'erreur quadratique moyenne  en fonction de degré de polynome")
plt.show()
```

```python
    polyreg1D = PolynomialFeatures(degree =8) # Transform X into X_Poly which contains X, X squared, X cubed, etc
    X_Poly = polyreg1D.fit_transform(X_train_df)
    regressor1 = LinearRegression() # Fit X_Poly in our Linear Regression model
    regressor1.fit(X_Poly,y_train)
    # Predicting a new result
    y_pred = regressor1.predict(X_Poly)
    mse = mean_squared_error(y_train, y_pred)
    print(mse)
    # Predicting a new result
    Xt_Poly = polyreg1D.fit_transform(X_test_df)
    y_pred1 = regressor1.predict(Xt_Poly)
    mset = mean_squared_error(y_test, y_pred1)
    print(mset)
```

```python
#graphique avec les échelles carrées
x_axis = np.arange(0, 100, 0.1)
Xt_Poly = polyreg1D.fit_transform(x_axis.reshape(-1,1))
y_pred1 = regressor1.predict(Xt_Poly)
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(X_train,y_train,'o')
ax.plot(x_axis,y_pred1)
ax.set_xlabel("Total CPU")
ax.set_ylabel("Temperature de CPU Package")
plt.title("visualisation")
```
